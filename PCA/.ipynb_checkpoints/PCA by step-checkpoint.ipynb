# Create Data for PCA

```python
import numpy as np
import math
np.random.seed(4294967295) # random seed for consistency

mu_vec1 = np.array(30 * [0]) #number of samples
cov_mat1 = np.identity(30) # number of samples
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20) # number of features
# class1_sample *= 10
```

```python
%pylab inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
plt.rcParams['legend.fontsize'] = 10   
ax.plot(class1_sample[0,:], class1_sample[1,:], class1_sample[2,:], 'o', markersize=8, color='blue', alpha=0.5, label='class1')

plt.title('Samples for class 1 (first 3 dim)')
ax.legend(loc='upper right')

plt.show()
```

```python
all_samples = np.copy(class1_sample.T)
print 'sample data'
print np.shape(all_samples)
```

# PCA

```python
from sklearn.decomposition import PCA as sklearnPCA
```

```python
print shape(all_samples)
pca = sklearnPCA(n_components=10)
fitted = pca.fit(all_samples)
trans = pca.transform(all_samples)
print shape(trans)
```

```python
plt.plot(trans[:,0],trans[:,1], 'o', markersize=7, color='blue', alpha=0.5, label='class1')
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.xlim([-8,8])
plt.ylim([-8,8])
plt.legend()
plt.title('Transformed samples without dropout rate')

plt.show()
```

# Apply Drop Out Rate ($e^{-\lambda\mu^2}$) / Gaussian Noise

I believe the proper order is:
1. generate low dim data
2. add gaussian noise to low dim
3. project back to high dim and apply drop out rate

```python
lmbda = -.5
mu = 2
p_0 = math.e**(lmbda * mu**2)
p_0
```

```python
def drop_out(p):
    return np.random.binomial(1, p, 1)[0]
```

```python
#s = np.random.binomial(1, p_0, 100000000)
#print sum(s)/100000000.0
```

```python
t = all_samples
shape(t)[0]
```

```python
#gaussian noise
noise_m = .1 * np.random.randn(shape(all_samples)[0], shape(all_samples)[1]) 
```

# dropped out data

```python
noise_drop_matrix = np.copy(all_samples) + noise_m
```

```python
count = 0
for (i,j), value in numpy.ndenumerate(noise_drop_matrix):
    if (drop_out(p_0) == 1):
        count += 1
        if (i == 0):
            print i, j
        noise_drop_matrix[i,j] = 0
print count
```
